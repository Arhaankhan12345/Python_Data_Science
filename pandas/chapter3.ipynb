```python
cols_to_drops=['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drops,inplace=True)
```

```python
print(df.columns.to_list())
```

```python
df.rename(columns={
'OdName':'Country','AreaName':'Continent',
'RegName':'Region','DevName':'Status'},
inplace=True)
```

```python
df.head(3)
```

```python
years=list(range(1980,2014))
print(years)
```

```python
df[years].sum()
```

```python
df[years].sum(axis=1)
```

```python
df['Total']=df[years].sum(axis=1)
```

```python
df.head(3)
```

```python
df.set_index('Country', inplace=True)
```

```python
df.sort_values(by=1980, ascending=False).head(25)[1980].plot(kind='bar', figsize = (20,5))
```

```python
df.sort_values('Total',ascending=False,inplace=True)
df.head()
```

```python
df.loc['Japan',years].plot(kind='line', style='ro-',title='japan Immigration Trend',figsize=(15,5))
```
